```python
df = pd.read_csv('gene_expression.csv')
```

```python
print(gene_expression_df.head())
```

```python
# Basic statistics for gene expression data
gene_exp_mean = df["Gene_Expression"].mean()
gene_exp_A_mean = df["Gene_A_Expression"].mean()
gene_exp_B_mean = df["Gene_B_Expression"].mean()
print("Mean:")
print("Gene_Expression:", gene_exp_mean)
print("Gene_A_Expression:", gene_exp_A_mean)
print("Gene_B_Expression:", gene_exp_B_mean)
```

```python
gene_exp_median = df["Gene_Expression"].median()
gene_exp_A_median = df["Gene_A_Expression"].median()
gene_exp_B_median = df["Gene_B_Expression"].median()
print("Median:")
print("Gene_Expression:", gene_exp_median)
print("Gene_A_Expression:", gene_exp_A_median)
print("Gene_B_Expression:", gene_exp_B_median)
```

```python
gene_exp_std = df["Gene_Expression"].std()
gene_exp_A_std = df["Gene_A_Expression"].std()
gene_exp_B_std = df["Gene_B_Expression"].std()
print("\nStandard Deviation:")
print("Gene_Expression:", gene_exp_std)
print("Gene_A_Expression:", gene_exp_A_std)
print("Gene_B_Expression:", gene_exp_B_std)
```

```python
# Convert all values to numeric, forcing errors to NaN (if any)
gene_expression_df = gene_expression_df.apply(pd.to_numeric, errors='coerce')

# Verify the data types to ensure it's numeric
print(gene_expression_df.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of gene expression levels
plt.figure(figsize=(10, 6))
sns.histplot(gene_expression_df.values.flatten(), bins=50, kde=False)
plt.title('Distribution of Gene Expression Levels')
plt.xlabel('Expression Level')
plt.ylabel('Frequency')
plt.savefig('histogram.png')
plt.show()

```

```python
# Density plot of gene expression levels
plt.figure(figsize=(10, 6))
sns.kdeplot(gene_expression_df.values.flatten(), shade=True)
plt.title('Density Plot of Gene Expression Levels')
plt.xlabel('Expression Level')
plt.ylabel('Density')
plt.savefig('density_plot.png')
plt.show()
```

```python
# Boxplot of gene expression levels
plt.figure(figsize=(12, 6))
sns.boxplot(data=gene_expression_df)
plt.title('Boxplot of Gene Expression Levels Across Samples')
plt.xlabel('Samples')
plt.ylabel('Expression Level')
plt.savefig('boxplot.png')
plt.show()
```

```python
# Violin plot of gene expression levels
plt.figure(figsize=(12, 6))
sns.violinplot(data=gene_expression_df)
plt.title('Violin Plot of Gene Expression Levels Across Samples')
plt.xlabel('Samples')
plt.ylabel('Expression Level')
plt.savefig('violin_plot.png')
plt.show()
```

```python
# Scatter plot of two selected genes (replace 'Gene1' and 'Gene2' with actual gene names)
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(gene_expression_df['Gene_A_Expression'], gene_expression_df['Gene_B_Expression'], alpha=0.5)
plt.title('Scatter Plot Between Gene1 and Gene2 Expression Levels')
plt.xlabel('Gene_A_Expression')
plt.ylabel('Gene_B_Expression')
plt.savefig('scatterplot.png')
plt.show()
```
